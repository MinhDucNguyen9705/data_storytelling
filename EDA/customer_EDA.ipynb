```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
customer = pd.read_csv('../data_cleaning/customer_cleaned.csv')
ticket = pd.read_csv('../data_cleaning/ticket_cleaned.csv')
film = pd.read_csv('../data_cleaning/film_cleaned.csv')
```

```python
customer_and_ticket = customer.merge(ticket, how='inner')
customer_and_ticket
```

```python
customer_and_ticket.columns
```

```python
customer_and_ticket[['job','industry', 'age']]
```

```python
customer_and_ticket['industry'] = customer_and_ticket['industry'].fillna('No work')
```

```python
customer_and_ticket
```

```python
customer_and_ticket['saledate'] = pd.to_datetime(customer_and_ticket['saledate'])
customer_and_ticket = customer_and_ticket.sort_values(by=['saledate'], ascending=True)
customer_and_ticket['saledate']
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=customer, x='district').set(title='Number of customers in each district')
plt.xticks(rotation=50)
plt.show()
```

```python
saledate_and_district = customer_and_ticket[['customerid', 'saledate', 'district']]
saledate_and_district
```

```python
districts =  np.array(saledate_and_district['district'].unique())
```

```python
counts_per_day = saledate_and_district.groupby(['saledate', 'district']).size().reset_index(name='count')
plt.figure(figsize=(20, 6))
sns.lineplot(x='saledate', y='count', hue='district', data=counts_per_day)
plt.title('Value Counts per Day of Each District')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks([])
plt.show()
```

```python
saledate_industry_district = customer_and_ticket[['customerid', 'saledate', 'industry', 'district']]
```

```python
plt.figure(figsize=(20, 5))
data = np.array(customer_and_ticket['industry'].value_counts())
plt.pie(data, labels=saledate_industry_district['industry'].value_counts().index, autopct="%.0f%%")
plt.show()
```

```python
saledate_industry_district = saledate_industry_district[saledate_industry_district['industry'] != 'No work']
```

```python
plt.figure(figsize=(20,5))
industry_counts_per_day = saledate_industry_district[['customerid', 'saledate', 'industry']].groupby(['saledate', 'industry']).size().reset_index(name='count')
print(industry_counts_per_day)
sns.lineplot(x='saledate', y='count', hue='industry', data=industry_counts_per_day)
plt.show()
```

```python
plt.figure(figsize=(20,5))
industry_counts_per_day = saledate_industry_district[['customerid', 'saledate', 'industry']].groupby(['saledate', 'industry']).size().reset_index(name='count')
print(industry_counts_per_day)
sns.relplot(data=industry_counts_per_day, x='saledate', y='count', kind='line', row='industry', hue='industry', height=3, aspect=3)
plt.xticks([])
plt.show()
```

Analysis about the number of "No work" customers

```python
nowork_customer_and_ticket = customer_and_ticket[customer_and_ticket['industry'] == 'No work']
nowork_customer_and_ticket
```

```python
nowork_date_age_ticketprice = nowork_customer_and_ticket[['customerid', 'saledate', 'age', 'ticket price', 'total']]
```

```python
sns.histplot(data=nowork_customer_and_ticket, x='age', color='skyblue')
plt.title('Number of customers at each age')
plt.show()
```

```python
#total sales per day of no work customers
nowork_sales_per_day = nowork_customer_and_ticket.groupby(['saledate'])['total'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(data=nowork_sales_per_day)
plt.xticks([])
plt.title('Total sales per day of "no work" customers')
plt.show()
```
