```python
import pandas as pd 
import numpy as np 
```

```python
ticket = pd.read_csv('ticket.csv')
```

```python
ticket.head()
```

```python
ticket.info()
```

```python
ticket['row'] = ticket['slot'].apply(lambda x: x[0])
ticket['seat'] = ticket['slot'].apply(lambda x: int(x[1:]))
```

```python
ticket.head()
```

```python
ticket.isna().sum()
```

```python
ticket.dropna(axis=0, inplace=True)
```

```python
ticket['hour'] = ticket['time'].apply(lambda x: x[:x.index(':')])
```

```python
ticket['saledate'] = pd.to_datetime(ticket['saledate'], format='%m/%d/%Y')
ticket['date'] = pd.to_datetime(ticket['date'], format='%m/%d/%Y')
ticket['time'] = pd.to_datetime(ticket['time'], format='%H:%M').dt.time
```

```python
ticket['purchased in advance'] = ticket['date']-ticket['saledate']
```

```python
ticket.head()
```

```python
ticket[['cashier', 'slot', 'room', 'slot type', 'ticket type', 'popcorn', 'row', 'seat', 'hour']] = ticket[['cashier', 'slot', 'room', 'slot type', 'ticket type', 'popcorn', 'row', 'seat', 'hour']].astype('category')
ticket['total'] = ticket['total'].astype('int')
ticket['hour'] = ticket['hour'].cat.reorder_categories(new_categories=list(str(i) for i in range(7, 24)), ordered=True)
ticket['seat'] = ticket['seat'].cat.reorder_categories(new_categories=list(range(1, 20)), ordered=True)
# ticket['hour'] = ticket['hour'].cat.reorder_categories(new_categories=list(chr(i) for i in range(7, 24)), ordered=True)
```
